```python
import json
import os
import cv2
```

```python
BASE_IM_PATH = 'dataset/real_data'
RAW_LABELS_PATH = 'dataset/real_data/labels.json'
LABELS_PATH = 'dataset/real_data/labels_new.json'
RAW_IMAGES_PATH = 'dataset/real_data/raw_images'
BASE_CROP_SIZE = 64
```

```python
with open(LABELS_PATH) as labels:
    raw_data = json.load(labels)
raw_data
```

```python
trash = []
print(len(raw_data.keys()))
for key in raw_data.keys():
    if raw_data[key]['regions']['0']['shape_attributes']['width'] > 64 \
    or raw_data[key]['regions']['0']['shape_attributes']['height'] > 64:
        trash.append(key)
for key in trash:
    raw_data.pop(key)
print(raw_data)
```

```python
data = []
for bg_type in os.listdir(BASE_IM_PATH):
    if not bg_type.endswith('.json') and not bg_type.endswith('.jpg'):
        for dist in os.listdir(os.path.join(BASE_IM_PATH, bg_type)):
            for name in os.listdir(os.path.join(BASE_IM_PATH, bg_type, dist)):
                item = {}
                item['name'] = name
                item['dist'] = float(dist.replace('_', '.'))
                item['mult'] = 1
                data.append(item)
print(data)
```

```python
raw_data_names = [x.split('.')[0] + '.jpg' for x in raw_data.keys()]
print(raw_data_names)
```

```python
data = [x for x in data if x['name'] in raw_data_names]
print(data)
```

```python
for item in data:
    for k in raw_data.keys():
        if k.startswith(item['name']):
            key = k
    print(key)
    x = raw_data[key]['regions']['0']['shape_attributes']['x']
    y = raw_data[key]['regions']['0']['shape_attributes']['y']
    width = raw_data[key]['regions']['0']['shape_attributes']['width']
    height = raw_data[key]['regions']['0']['shape_attributes']['height']
    item['x_c'] = x + width // 2
    item['y_c'] = y + height // 2
print(data)
```

```python
def get_crop_roi(x_ctr, y_ctr):
    crop_size = BASE_CROP_SIZE
        
    left = int(x_ctr - crop_size / 2)
    right = int(x_ctr + crop_size / 2)
    top = int(y_ctr - crop_size / 2)
    bottom = int(y_ctr + crop_size / 2)
    return left, right, top, bottom
```

```python
im_names = [x['name'] for x in data]
for name in im_names:
    im = cv2.imread(os.path.join(RAW_IMAGES_PATH, name))
    for item in data:
        if item['name'] == name:
            i = item
    left, right, top, bottom = get_crop_roi(i['x_c'], i['y_c'])
    im = im[top:bottom, left:right, :]
    cv2.imwrite(os.path.join(BASE_IM_PATH, name), im)
```

```python
with open(LABELS_PATH, 'w+') as labs:
    json.dump({'data': data}, labs)
```
